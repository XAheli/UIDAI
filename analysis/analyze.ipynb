```python


# Create a mapping dictionary for state name corrections
state_corrections = {
    # West Bengal variations
    'WEST BENGAL': 'West Bengal',
    'Westbengal': 'West Bengal',
    'West bengal': 'West Bengal',
    'west Bengal': 'West Bengal',
    'West  Bengal': 'West Bengal',
    'West Bangal':'West Bengal',
    'WESTBENGAL': 'West Bengal',
    'West Bengli':'West Bengal',
    # Odisha variations (Orissa is the old name)
    'Orissa': 'Odisha',
    'ODISHA': 'Odisha',
    'odisha': 'Odisha',
    
    # Andhra Pradesh variations
    'andhra pradesh': 'Andhra Pradesh',
    'Tamilnadu':'Tamil Nadu',
    # Pondicherry is now Puducherry
    'Pondicherry': 'Puducherry',
    
    # Andaman & Nicobar Islands variations
    'Andaman & Nicobar Islands': 'Andaman and Nicobar Islands',
    
    # Jammu & Kashmir variations
    'Jammu & Kashmir': 'Jammu and Kashmir',
    
    # Daman and Diu variations
    'Daman & Diu': 'Daman and Diu',
    
    # Dadra and Nagar Haveli variations
    'Dadra & Nagar Haveli': 'Dadra and Nagar Haveli',
    
    # Uttaranchal is now Uttarakhand
    'Uttaranchal': 'Uttarakhand',
    'Jammu And Kashmir':'Ladakh',
    # Chhattisgarh spelling
    'Chhatisgarh': 'Chhattisgarh',

    'Dadra and Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'Daman and Diu':'Dadra and Nagar Haveli and Daman and Diu',
    'Daman and Diu':'Dadra and Nagar Haveli and Daman and Diu',
    'Dadra and Nagar Haveli':'Dadra and Nagar Haveli and Daman and Diu',
    # Merge Dadra and Nagar Haveli and Daman and Diu (merged in 2020)
    'Dadra and Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'Daman and Diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'The Dadra And Nagar Haveli And Daman And Diu': 'Dadra and Nagar Haveli and Daman and Diu'
}

# Apply the corrections
df['state'] = df['state'].replace(state_corrections)

# Verify the results
print(df['state'].value_counts())
```

```python
invalid_entries = ['Darbhanga', 'Puttenahalli','BALANAGAR','100000','Jaipur','Madanapalle','Raja Annamalai Puram','Nagpur']
df = df[~df['state'].isin(invalid_entries)]
```
